# Filtrando Dados


```python
# Filtrando pokemons do tipo grama e veneno com mais de 75 de HP
# and em pandas &
# or em pandas |
poke.loc[(poke['Type 1'] == 'Grass') & (poke['Type 2'] == 'Poison') & (poke['HP'] > 75)]
```

```python
# Filtrando todos os pokemons que contem MEGA em seus nome
poke.loc[poke['Name'].str.contains('Mega')]
```

```python
# Filtrando todos os pokemons que NAO contem MEGA em seus nome
poke.loc[~poke['Name'].str.contains('Mega')]
```

```python
# Filtrando pokemons utilizando regex
import regex as re
poke.loc[poke['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]
```

```python
# Filtrano com regex pokemons que contenham as letras 'Pi'
poke.loc[poke['Name'].str.contains('^pi[a=z]*', flags=re.I, regex=True)]
```

# Agregando estatistica utilizando (groupyby)

```python
# Recarregando o data frame 
poke = pd.read_csv('pokemon.csv')


# Mostrando a media de defesa dos pokemons 
'''poke.groupby(poke['Type 1']).mean().sort_values('Defense', ascending=False)'''


# Mostrando a media dos pokemons com maior ataque
'''poke.groupby(poke['Type 1']).mean().sort_values('Attack', ascending=False)'''


# Somando todos os tipos de pokemon do tipo 1, no caso nao fara sentido neste
# contexto porem vamos anotar kk
'''poke.groupby(poke['Type 1']).sum().sort_values('Total', ascending=False)'''


# Contando todos os tipos de pokemons
'''poke.groupby(poke['Type 1']).count()'''

# Adicionando uma nova coluna para contarmos com melhor visualizacao
poke['Quantidade'] = 1

poke.groupby(poke['Type 1']).count()['Quantidade']


# Plotando um grafico com a coluna Quantidade
# Necessario import numpy
grafico = poke.groupby(poke['Type 1']).count()['Quantidade']
grafico.sort_values(ascending=False).plot.bar(title='Pokemons e suas quantidades')
```
